```python
import matplotlib.pyplot as plt
import numpy as np
import json
import pandas as pd
```

```python
with open('static_dictionary_Firefox_Lsvm11.json') as file:
    data = json.load(file)
#     print(data)
    
```

```python
ratios = []

for lexicon, words in data['Severe Lexicons'].items():
      for word, ratio_value in words.items():
        ratios.append(ratio_value) 
        
# print("'ratio':", ratios)
```

```python
data = {
    'severe_ratios':ratios
}

ratios = data['severe_ratios']
print(ratios)
```

```python


# Calculate mean
mean_value = sum(ratios) / len(ratios)

# Sort the list for median and min/max calculations
sorted_ratios = sorted(ratios)

# Calculate median
median_index = len(sorted_ratios) // 2
if len(sorted_ratios) % 2 == 0:
  # If even number of elements, median is average of middle two values
  median_value = (sorted_ratios[median_index - 1] + sorted_ratios[median_index]) / 2
else:
  # If odd number of elements, median is the middle value
  median_value = sorted_ratios[median_index]

# Minimum and maximum values
min_value = sorted_ratios[0]
max_value = sorted_ratios[-1]

# Print results
print("Mean:", mean_value)
print("Median:", median_value)
print("Min:", min_value)
print("Max:", max_value)

# print(f"Mean: {mean_value:.6f}")
# print(f"Median: {median_value:.6f}")
# print(f"Maximum: {max_value:.6f}")
# print(f"Minimum: {min_value:.6f}")

# Create a bar plot
plt.figure(figsize=(8, 6))
plt.bar(["Mean", "Median", "Max", "Min"], [mean_value, median_value, max_value, min_value], color=["b", "g", "r", "purple"])
plt.xlabel("Statistics")
plt.ylabel("Values")
plt.title("Mean, Median, Max, and Min Values")
plt.grid(axis="y")
plt.show()
```
